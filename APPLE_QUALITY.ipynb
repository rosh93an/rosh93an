```python
data["Quality"].value_counts()
```

```python
#feature and target
X=data.drop("Quality",axis=1)
Y=data["Quality"]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1) # #test size means 30% for testing 
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression #since logistic regression is statistical method
#designed for binary classification problems ,where the out come has two possible classes good and bad
```

```python
log=LogisticRegression()
```

```python
log.fit(X_train,Y_train)
```

```python
log.score(X_train,Y_train) #it checks the mean accuracy 
```

```python
Y_pred=log.predict(X_train)
```

```python
Y_pred
```

```python
log.score(X_test,Y_test)
```

```python
Y_predtest=log.predict(X_test)
```

```python
Y_predtest
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(Y_test,Y_predtest)) 
```

```python
#since accuracy is 0.76  via logistic regression ,classification report  so prefered adaboostclassifier  for better accuracy 
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada=AdaBoostClassifier(n_estimators=100)
```

```python
ada.fit(X_train,Y_train)
```

```python
ada.score(X_train,Y_train)
```

```python
Y_pred1=ada.predict(X_train)
```

```python
Y_pred1
```

```python
ada.score(X_test,Y_test)
```

```python
Y_predtest1=ada.predict(X_test)
```

```python
Y_predtest1
```

```python
print(classification_report(Y_test,Y_predtest1))
```

```python
#after adaboost classifier ,accuracy is still 0.78 but for better prediction model testing shoud be more then 0.80 so prefer Gradient boosting 
```

```python
#Boosting focuses on improving mistakes made by weak learners (models that perform slightly better than random chance).
#It sequentially adds up these weak learners, refining predictions with each step.
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb=GradientBoostingClassifier(n_estimators=100)
```

```python
gb.fit(X_train,Y_train)
```

```python
gb.score(X_train,Y_train)
```

```python
y_predgb=gb.predict(X_train)
```

```python
y_predgb
```

```python
gb.score(X_test,Y_test)
```

```python
y_predtesgbt=gb.predict(X_test)
```

```python
y_predtesgbt
```

```python
print(classification_report(Y_test,y_predtesgbt))
```

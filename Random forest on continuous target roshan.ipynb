```python
dfnew=pd.concat([df_num,df_cat],axis=1)
```

```python
dfnew
```

```python
x=dfnew.drop("actual",axis=1)
y=dfnew["actual"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
#we are using regressor instead of classifier because my target data is continuous
```

```python
rf=RandomForestRegressor(n_estimators=50,max_depth=10)
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)
```

```python
y_pred=rf.predict(x_train)
```

```python
y_pred
```

```python
rf.score(x_test,y_test)
```

```python
Y_predtest=rf.predict(x_test)
```

```python
Y_predtest
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse=mean_squared_error(y_test,Y_predtest)
print(mse)
```

```python
r2score=r2_score(y_test,Y_predtest)
print(r2score)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Diabetes.csv")
```

```python
for i in df:
    print(i,":",df[i].unique())
    print("*******************************************************************************")
```

```python
df.isnull().sum()
```

```python
sns.countplot(data=df,x="TenYearCHD")
plt.show()
```

```python
List2=["education","cigsPerDay","BPMeds","totChol","heartRate","glucose","BMI"]
for i in List2:
    print(i,":",df[i].mean())
    print("*******************************")
```

```python
df["education"].fillna(1.97,inplace=True)
df["cigsPerDay"].fillna(9,inplace=True)
df["BPMeds"].fillna(0.02,inplace=True)
df["totChol"].fillna(236.72,inplace=True)
df["heartRate"].fillna(75.87,inplace=True)
df["glucose"].fillna(81.96,inplace=True)
df["BMI"].fillna(25.80,inplace=True)
```

```python
#Feature & Target

x=df.drop("TenYearCHD",axis=1)
y=df["TenYearCHD"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()  
```

```python
log.fit(x_train,y_train)
```

```python
log.score(x_train,y_train)
```

```python
y_pred=log.predict(x_train)
```

```python
y_pred
```

```python
log.score(x_test,y_test)
```

```python
y_predtest=log.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_predtest))
```

```python
tn,fp,fn,tp=confusion_matrix(y_test,y_predtest).ravel()
```

```python
tn #actual no predicted no
```

```python
fn #actual yes predicted no
```
